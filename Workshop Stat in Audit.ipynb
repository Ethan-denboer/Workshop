### Stap 4: Analyse
Bovenstaande verkenning is natuurlijk heel inzichtelijk en ongetwijfeld enorm nuttig.
Het op het oog vergelijken van de verschillende verdelingen van variabelen tussen categorien is echter niet altijd even makkelijk.
Hiervoor kunnen we een statische toets gebruiken waarbij de verdelingen onderling worden getoetst op significante afwijkingen van elkaar.

De bekendste toets is de Student-t Test. Deze is enorm krachtig en veel gebruikt, maar ook misbruikt.
Deze toets mag in principe alleen gebruikt worden voor het vergelijken van normaal verdeelde populaties.

Onze 'real life' data hierboven is veel, maar niet normaal verdeeld.

Hiervoor vallen we terug op non-parametrische toetsen zoals de Mann-Whitney.

````
## Opdracht. Vervang bij Selectie de code rechts van de '=' met een getal kleiner dan het maximum van de histogram
## Draai de cel

## Bonus: Vervang de waarden rechts van de '=' bij variable en per met een variabele en categorie naar keuze
## PAS OP! de per moet een categorie (object) zijn, anders werkt het waarschijnlijk niet.
## PAS OP! de per kan niet meer dan 2 categorien bevatten voor deze toets

Selectie = max(dataSet.Cost) 
variabele = 'Cost'
per = 'Jaar'

## Achtergrondcode.
data = dataSet
cat1 = pd.unique(dataSet[per])[0]
cat2 = pd.unique(dataSet[per])[1]

x = data[variabele][data[per] == cat1]
y = data[variabele][data[per] == cat2]
stat, p = mannwhitneyu(x, y)
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p < 0.05:
    print('De verdelingen zijn significant verschillend')
else:
    print('De verdelingen zijn niet verschillend')
````
